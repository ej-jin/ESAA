## HW1-2. dislikes 수 대비 likes 수

```python
pure_data=row_concat[row_concat['likes']>0]
more_pure_data=pure_data[pure_data['dislikes']>0]
l_d=more_pure_data['dislikes']/more_pure_data['likes']
more_pure_data.head()
```

```python
more_pure_data["l_d"]=l_d
```

```python
sort_by_l_d_row_concat=more_pure_data.sort_values(by="l_d",ascending=False)
sort_by_l_d_row_concat
```

```python
l_d_by_category_id = sort_by_l_d_row_concat.groupby('category_id')
avg_l_d_by_category_id = l_d_by_category_id.l_d.mean()
avg_l_d_by_category_id
```

```python
ax = plt.subplots() 
ax = sns.regplot(x="views",y="likes",data=row_concat)
# tips 데이터에서 x축을 total_bill, y축을 tip으로 regplot 그리기 
ax.set_title('views VS likes') 
ax.set_xlabel('views') 
ax.set_ylabel('likes')
```

```python
outlier_likes1=row_concat[row_concat["likes"]>3500000]
outlier_likes2=outlier_likes1[outlier_likes1["views"]<150000000]
outlier_likes2
```

#### 결과: 조회수 대비 유난히 like수가 많은 영상은 특이하게도 category가 모두 10(music)이였다. "Childish Gambino - This Is America","BTS-'FAKE LOVE'"의 뮤직비디오였다.  팬덤 파워를 느낄 수 있는 outlier였다.

US의 category_id별 조회수(views)결과:  
category_id별 조회수 확인 결과 10:Music 분야가 가장 높게 나타났다.  
평균 막대그래프를 그려보니 조회수 평균이 가장 높게 나타난 분야 또한 10:Music이었다.

CA의 category_id별 조회수(views)결과:

## HW1-4. 코멘트와 조회수의 관계

```python
data2 = CA.loc[:,['category_id','views', 'comment_count']]
data2.head()
```

```python
data2.sort_values(by="views",ascending=False)
```

```python
data2.sort_values(by="comment_count",ascending=False).head()
```

결론: 조회수가 가장 많은 영상에 코멘트가 가장 많이 달렸을 것으로 예상했다.  
     그러나 우리의 예상과는 달리 조회수가 많다고 코멘트가 가장 많이 달리지 않았다는 분석 결과를 확인할 수 있었다.

### 코멘트와 조회수의 관계 산점도

```python
# CA 데이터
ax = plt.subplots() 
ax = sns.regplot(x='views', y='comment_count', data=CA, fit_reg=True)
ax.set_title('Scatterplot of views vs comment_count') 
ax.set_xlabel('views') 
ax.set_ylabel('comment_count') 
```

```python
# data2 데이터: CA데이터에서 views, comment_count 만 추출한 데이터
ax = plt.subplots() 
ax = sns.regplot(x='views', y='comment_count', data=data2, fit_reg=True)
ax.set_title('Scatterplot of views vs comment_count') 
ax.set_xlabel('views') 
ax.set_ylabel('comment_count') 
```

코멘트수와 조회수 간의 관계를 산점도를 그려 확인해본 결과 하위 그룹에 밀집되어 있어서 확인이 어렵다.  
따라서 밀집된 부분의 데이터를 별도로 추출해보기로 하였다.

```python
re_data = data2.sort_values(by="views",ascending=True).head(100)
print(re_data.head(10))
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='views', y='comment_count', data=re_data, fit_reg=True)
ax.set_title('Re Scatterplot of views vs comment_count') 
ax.set_xlabel('views') 
ax.set_ylabel('comment_count') 
```

결론: 코멘트수와 조회수 간의 관계에서 하위 그룹에 밀집된 데이터들의 산점도를 그려본 결과 약한 양의 상관관계를 보이는 것을 확인했다.

## HW1-5. category_id별 가장 Hot한 분야

```python
ids = CA.category_id.unique() 
print(sorted(ids))
print(len(sorted(ids)))
```

```python
CA_01=CA.loc[:, ['category_id', 'views', 'likes', 'dislikes', 'comment_count']]
CA_01.head()
```

```python
group1 = CA_01.groupby('category_id').views.sum()
print(group1.sort_values(ascending=False))
# 조회수는 24:Entertainment 분야가 가장 많다.
```

```python
group2 = CA_01.groupby('category_id').likes.sum()
print(group2.sort_values(ascending=False))
# 좋아요수는 10:Music 분야가 가장 많다.
```

```python
group3 = CA_01.groupby('category_id').dislikes.sum()
print(group3.sort_values(ascending=False))
# 싫어요수는 24:Entertainment 분야가 가장 많다.
```

```python
group4 = CA_01.groupby('category_id').comment_count.sum()
print(group4.sort_values(ascending=False))
# 코멘트수는 24:Entertainment 분야가 가장 많다.
```

```python
groups = pd.DataFrame([group1, group2, group3, group4])
groups
```

결론: category_id별로 분석한 결과 조회수, 코멘트수는 24:Entertainment 분야가 가장 많았다.  
또한 좋아요를 가장 많이 받은 분야는 10:Music이며, 싫어요를 가장 많이 받은 분야는 조회수가 가장 많았던 24:Entertainment였다.

```python
ax = plt.subplots() 
ax = sns.countplot('category_id', data=CA_01) 
ax.set_title('Frequency of category_id') 
ax.set_xlabel('category_id') 
ax.set_ylabel('Frequency') 
```
